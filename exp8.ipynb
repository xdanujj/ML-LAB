```python
# McCulloch-Pitts Neuron Implementation for Logic Gates

# Step 1: Define the MP Neuron model
def MP_Neuron(inputs, weights, threshold):
    weighted_sum = sum(i * w for i, w in zip(inputs, weights))
    output = 1 if weighted_sum >= threshold else 0
    return output


# Step 2: Implement basic logic gates

# AND Gate
def AND_gate(x1, x2):
    weights = [1, 1]
    threshold = 2
    return MP_Neuron([x1, x2], weights, threshold)

# OR Gate
def OR_gate(x1, x2):
    weights = [1, 1]
    threshold = 1
    return MP_Neuron([x1, x2], weights, threshold)

# NOT Gate (single input)
def NOT_gate(x):
    weight = [-1]
    threshold = 0
    return MP_Neuron([x], weight, threshold)


# Step 3: Test all gates
print("AND Gate Truth Table")
for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(f"{x1} AND {x2} = {AND_gate(x1, x2)}")

print("\nOR Gate Truth Table")
for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(f"{x1} OR {x2} = {OR_gate(x1, x2)}")

print("\nNOT Gate Truth Table")
for x in [0, 1]:
    print(f"NOT {x} = {NOT_gate(x)}")
```
